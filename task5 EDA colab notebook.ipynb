```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

df = pd.read_csv("/content/titanic.csv")

print("First 5 rows:")
print(df.head())
```

```python
print("\nDataset Info:")
df.info()
```

```python
print("\nStatistical Description:")
print(df.describe(include='all'))
```

```python
print("\nUnique Value Counts:")
print(df.nunique())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nCategorical Value Counts:")
print(df['Embarked'].value_counts())
```

```python
# Histograms for all numeric columns
df.hist(bins=25, figsize=(12, 8), color='steelblue', edgecolor='black')
plt.suptitle("Histograms of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
# Boxplots to detect outliers in Age and Fare
sns.boxplot(data=df[['Age', 'Fare']], palette="Set2")
plt.title("Boxplot of Age and Fare")
plt.show()
```

```python
# Countplot: Passenger class distribution
sns.countplot(data=df, x='Pclass', palette='pastel')
plt.title("Passenger Class Distribution")
plt.show()
```

```python
# Countplot: Survival count by gender
sns.countplot(data=df, x='Sex', hue='Survived', palette='muted')
plt.title("Survival by Gender")
plt.show()
```

```python
# Pairplot to examine pairwise relationships
sns.pairplot(df.dropna(), hue='Survived', palette='husl')
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

```python
# Scatterplot: Age vs Fare colored by Survival
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='coolwarm')
plt.title("Age vs Fare (colored by Survival)")
plt.show()
```

```python
# Heatmap (correlation)
plt.figure(figsize=(10, 6))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```
